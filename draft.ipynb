```python
league_rank = pd.read_csv('./data/proc/league_player_list_with_ranks.csv')#.set_index('ID')
league_stat_cast = pd.read_csv('./data/proc/league_player_list_with_statcast.csv')#.set_index('ID')
league_keepers = pd.read_csv('data/proc/league_keeper_list.csv')#.set_index('ID')
league_rank = league_rank[league_rank.ID.apply(lambda x: x  not in league_keepers.ID.values)]
league_stat_cast = league_stat_cast[league_stat_cast.ID.apply(lambda x: x  not in league_keepers.ID.values)]
```

```python
league_rank.head(50)
```

```python
def filt_by_position(df, position):
    return df[ df['Position'].apply(lambda x: position.lower() in x.lower() ) ]
```

## 1B situation

```python
first_base_rank = filt_by_position(league_rank, '1B')
first_base_rank["value"] = first_base_rank.ADP_round - first_base_rank.Rank_round
first_base_rank = first_base_rank[first_base_rank.Rank_round <= 15]
first_base_rank.sort_values('value', ascending=False)
```
