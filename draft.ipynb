```python
def filt_by_position(df, position):
    return df[ df['Position'].apply(lambda x: position.lower() in x.lower() ) ]
```

## 2B situation

```python
second_base_rank = filt_by_position(league_rank, '2B')
second_base_rank["value"] = second_base_rank.ADP_round - second_base_rank.Rank_round
second_base_rank = second_base_rank[second_base_rank.Rank_round <= 15]
second_base_rank.sort_values('value', ascending=False)
```
